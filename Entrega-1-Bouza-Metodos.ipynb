# EJERCICIO TRIBISECCIÓN 
* Implementar en Python un método para aproximar raíces al estilo de bisección, pero que en vez de dividir el intervalo en 2 subintervalos lo divida en 3, y en cada iteración elija uno de los 3 subintervalos que contenga una raíz. 
* Usar dos condición de parada análoga a la del método de bisección.

```python
#Programa que aproxima una raíz de una función con el método de trisección.

#Mínimo de pasos -> parámetro "pasos".

#Parámetros de parada -> epsilon y "pasos" .

#Aceptar input de los valores del intervalo inicial "a" y "b".
import math as m 
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return m.cos(x)

def signo(y,z):
    return (f(y)*f(z))

     
def triseccion(a, b, e, pasos):
    i = 0
    while (b-a) >= e or i < pasos:
        # Divido el tramo en tres partes iguales
        p_medio = (b-a)/3
        # Puntos C y D
        c = a + p_medio
        d = b - p_medio
        # Segmento A-C
        if signo(c, a) < 0:
            b = c
            i += 1
        # Segmento C-D
        elif signo(d, c) < 0:
            a = c
            b = d
            i += 1
        # Segmento D-B
        elif signo(b, d) < 0:
            a = d
            i += 1
    return a

def graficar(a,b):
    x_min=a
    x_max=b
    eje_x = np.linspace(x_min,x_max)
    eje_y = [f(x) for x in eje_x]
    plt.plot(eje_x, eje_y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Gráfico de la función')
    plt.grid(True)
    plt.ylim(bottom=0)
    plt.show()
    return
```

```python
print(triseccion(-1.0,2.0,0.1,10))
graficar(-1.0,2.0)
```

```python
a = float(input("Ingrese el primer valor del rango en formato decimal.\n"))
b = float(input("Ingrese el segundo valor del rango en formato decimal.\n"))
e = float(input("Ingrese el umbral de error en formato decimal.\n"))
pasos = int(input("Ingrese las iteraciones deseadas.\n"))

res = triseccion(a, b, e, pasos)

graficar(a,b)

print(res)
```
